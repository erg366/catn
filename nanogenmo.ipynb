```python
import random
import markovify
import tracery
from tracery.modifiers import base_english
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
all_sentences = []

the_republic_text = open("the_republic.txt").read()
republic_sent = the_republic_text.split(".")

words = []
noun_chunks = []
entities = []
name_entities = []
for i, sent in enumerate(random.sample(republic_sent, 500)):
    all_sentences.append(sent + ".")
    if i % 100 == 0:
        print(i, len(republic_sent))
    doc = nlp(sent)
    words.extend([w for w in list(doc) if w.is_alpha])
    noun_chunks.extend(list(doc.noun_chunks))
    name_entities.extend(list(doc.ents))
```

```python
people = [e for e in name_entities if e.label_ == "PERSON"]
for entity in name_entities:
    if entity in people:
        print(entity)
    else:
        entities.append(entity)
```

```python
nouns = [w for w in words if w.pos_ == "NOUN"]
verbs = [w for w in words if w.pos_ == "VERB"]
past_tense_verbs = [w for w in words if w.tag_ == 'VBD']
adjs = [w for w in words if w.tag_ == "ADJ"]
advs = [w for w in words if w.pos_ == "ADV"]
people = [e for e in entities if e.label_ == "PERSON"]
locations = [e for e in entities if e.label_ == "LOC"]
times = [e for e in entities if e.label_ == "TIME"]
subjects = [chunk for chunk in noun_chunks if chunk.root.dep_ == 'nsubj']
objects = [chunk for chunk in noun_chunks if chunk.root.dep_ == 'dobj']
```

```python
chorus_rules = {
    "subject": [w.text for w in subjects],
    "object": [w.text for w in objects],
    "verb": [w.text for w in verbs],
    "adj": [w.text for w in adjs],
    "adv": [w.text for w in advs],
    "ent": [w.text for w in entities],
    "people": [w.text for w in people],
    "loc": [w.text for w in locations],
    "time": [w.text for w in times],
    "origin": "\n\n\n\n#strophe#\n\n#antistrophe#\n\n#epode#\n\n\n\n",
    "strophe": "#sentence#\n#sentence#\n#sentence#\n#sentence#",
    "antistrophe": "#sentence#\n#sentence#\n#sentence#\n#sentence#",
    "epode": "#sentence#\n#sentence#",
    "sentence": [
        "[adjA:#adj#]#adjA.capitalize# #ent#, #adjA# #ent#",
        "[verbA:#verb#]#verbA.capitalize# #object#, #verbA# #object#",
        "#object# of #loc#",
        "no #ent#, no #ent# #verb#",
        "#adv# on #ent# he lies",
        "and #ent#'s #adj# form, and #subject# behind"
    ]
}

chorus_grammar = tracery.Grammar(chorus_rules)
chorus_grammar.add_modifiers(base_english)
print(chorus_grammar.flatten("#origin#"))
```

```python
for i in range(10):
    print(chorus_grammar.flatten("#origin#"))
    paragraph = ""
    for i in range(10):
        paragraph += (all_text_gen.make_short_sentence(100,tries=100) + " ")
    print(paragraph + "\n")
    paragraph = ""
    for i in range(10):
        paragraph += (all_text_gen.make_short_sentence(100,tries=100) + " ")
    print(paragraph)
```
